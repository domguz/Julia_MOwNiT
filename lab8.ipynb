```julia
import Pkg; Pkg.add("FFTW")
using Plots
using FFTW
```

```julia
Fs = 1024;  

t = 0:1/(Fs-1):1
x = sin.(2*pi*t*200) + 2* sin.(2*pi*t*400)
```

```julia
plot(t,x)
```

```julia
y=fft(x)
sticks((abs.(fft(x))))
```

```julia
Fs = 1024;  

t = 0:1/(Fs-1):1
x = cos.(2*pi*t*100)
for i in range(1,size(x,1))
    x[i] += rand(1:5)/10
end
```

```julia
plot(t,x)
```

```julia
y=fft(x)
sticks(abs.(y))
```

```julia
for i in range(1,size(y,1))
    if abs.(y[i]) < 50
        y[i] = 0
    end
end
sticks(abs.(y))
```

```julia
r = ifft(y)
a = zeros(1024)
for i in range(1,size(y,1))
    a[i] = r[i].re
end
plot(t,a)
```

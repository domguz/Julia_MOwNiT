```julia
using Plots, Roots, ForwardDiff;
```

```julia
f1(x) = 2 * x - exp(-x)
f2(x) = x * exp(-x)
f3(x) = exp(x) - 1/(10 * x)^2
f4(x) = (x + 3)*(x - 1)^2
f5(x) = cos(x) - x
f6(x) = x^3
functions = [(f1,(-0.5,3)),(f2,(-1,6)),(f3,(0,1)),(f4,(-4,2)),(f5,(0,1)),(f6,(-1,1))]
```

```julia
for (f,p) in functions
    println("Funkcja to ",f," przedział to ",p,"\n")
    x = find_zero(f,p,Bisection(), verbose=true)
    println("x0 = ",x," f(x0) = ",f(x))
end
```

```julia
#Newton
for (f,p) in functions
    D(f) = x->ForwardDiff.derivative(f, float(x))
    println("Funkcja to ",f,"\n")
    x = find_zero((f1, D(f1)),0.5, Roots.Newton(),verbose=true)
    println("x0 = ",x,"f(x0) = ",f(x))
end
```

```julia
#Steffensen
for (f,p) in functions
    println("Funkcja to ",f,"n")
    x = find_zero(f, 0.5, Order2(), verbose=true)
    println("x0 = ",x,"f(x0) = ",f(x),"\n")
end
```

```julia
using DataFrames
functions1 = [f1,f2,f3,f4,f5,f6]
Bisec_iterations = [60,0,62,60,60,0]
Bisec_evaluations = [62,2,64,62,62,2]
Newton_iterations = [4,4,4,4,4,4]
Newton_evaluations = [9,9,9,9,9,9]
Stef_iterations = [4,8,17,27,5,29]
Stef_evaluations = [8,12,28,49,9,54]
df = DataFrame(:f => functions1,:Bisec_iter => Bisec_iterations, :Bisec_eval => Bisec_evaluations)
df1 = hcat(df, DataFrame(:Newton_iter => Newton_iterations, :Newton_eval => Newton_evaluations))
df2 = hcat(df1, DataFrame(:Stef_iter => Stef_iterations, :Stef_eval => Stef_evaluations))
```

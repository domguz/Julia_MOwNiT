```julia
#zad3-5
using DataFrames
using CSV
using Statistics
using Plots

input="lab2.csv"
data=DataFrame()
data=CSV.read(input, delim=";")
df=by(data,:VectorSize,avgVec = :VxVTime => mean, avgMat = :MxVTime => mean,
                         devVec = :VxVTime => std, devMat = :MxVTime => std)

p1 = scatter(df[:VectorSize],df[:avgVec],title="Czas wektorXwektor",yerr=df[:devVec],
                          xlabel="wielkosc wektorow",ylabel="czas [ms]",legend=false)

p2 = scatter(df[:VectorSize],df[:avgMat],title="Czas macierzXwektor",yerr=df[:devMat],
                 xlabel="wielkosc wektora i macierzy",ylabel="czas [ms]",legend=false)

plot(p1,p2,layout=2)
```
